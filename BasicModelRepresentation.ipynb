<p style="font-size: 20px;">I will use the motivating example of housing price prediction to fit a linear regression model. This lab will use a simple data set with only two data points: a house with 500 square feet (sqft) sold for 300,000 dollars and a house with 1000 square feet sold for 500,000 dollars. These two points will constitute our data or training set. In this lab, the units of size are 1000 sqft, and the units of price are 1000s of dollars.</p>


<p style="font-size: 20px;">I would like to fit a linear regression model through these two points so I can then predict the price for other houses, such as a house with 700 sqft.</p>

<h3  style="font-size: 30px;">Tools:</h3>

<ul <p style="font-size: 20px;">
    <li>Numpy : a popular library for scientific computing</li>
    <li>Matplotlib : a popular library for plotting data</li>
</ul>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h3 style="font-size: 30px;">Creating dataset</h3>


```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([0.5, 1.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

<h3 style="font-size: 30px;">Plotting the data</h3>

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

<h3 style="font-size: 30px;">Model Function</h3>

<p style="font-size: 20px;">The model function for linear regression (which maps from x to y) is represented as:<br>
f(x) = wx+b<br>
The formula above is how we can represent straight lines. Different values of 
w and b gives different straight lines on the plot.<br>
Let's try to get a better intuition for this through the code blocks below. I'll start with w=500 and b=500.</p>

```python
w = 500
b = 500
print(f"w: {w}")
print(f"b: {b}")
```

```python
print(x_train.shape[0])
```

<p style="font-size: 20px;">Now I will compute the y values and compare it with the given y values </p>

```python
def computeprediction(x,w,b):
    m = x.shape[0]
    # creating empyt numpy array to store predicted values
    yprediction = np.zeros(m)
    for i in range(m):
        yprediction[i] = w*x[i]+b
    return yprediction
    
```

```python
predictedvalues = computeprediction(x_train, w,b)
predictedvalues
```

<p style="font-size: 20px;">Now we compare the predicted y values with the actual given y values. I will plot both the predicted and actual y values</p>

```python
# Plot our model prediction
plt.plot(x_train, predictedvalues, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

<p style="font-size: 20px;">As we can see having w = 500 and b = 500 did not give the correct prediction. Since this is a very basic lab I will update the values of w and b manually rather than using gradient descent</p>

```python
w = 400
b = 100
```

```python
predictedvalues = computeprediction(x_train, w,b)
predictedvalues
```

```python
# Plot our model prediction
plt.plot(x_train, predictedvalues, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

<h3 style="font-size: 30px;">Prediction</h3>
<p style="font-size: 20px;">Now that we have a model, we can use it to make our prediction. Let's predict the price of a house with 1200 sqft. Since the units of x are in 1000's of sqft, x is 1.2.</p>

```python
x_i = 1.2
pricefor1200 = w * x_i + b    

print(f"${pricefor1200:.0f} thousand dollars")
```
